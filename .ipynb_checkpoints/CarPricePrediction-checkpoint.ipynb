#### Import Libraries and Data

```python
# Importing all required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import statsmodels.api as sm 
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
%matplotlib inline
from sklearn import metrics
```

```python
#Load the Dataset 
car = pd.read_csv('CarPrices.csv')
car.head()
```

```python
# User defined function to calucate Variance Inflation Factor
def vif_cal(input_data, dependent_col):
    vif_df = pd.DataFrame( columns = ['Var', 'Vif'])
    x_vars=input_data.drop([dependent_col], axis=1)
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]] 
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.OLS(y,x).fit().rsquared  
        vif=round(1/(1-rsq),2)
        vif_df.loc[i] = [xvar_names[i], vif]
    return vif_df.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False)
```

```python
# User defined function to calcualate Ordinary Least Squares and VIF
def calculateOLSandVIF(input_data):
    global X_train
    global lm
    if set(input_data).issubset(car.columns):
        car.drop(input_data, axis =1, inplace = True)
    if set(input_data).issubset(X_train.columns):
        X_train = X_train.drop(input_data,1)
    lm = sm.OLS(y_train,X_train).fit()
    print(lm.summary())
    return vif_cal(input_data=car, dependent_col="price") 
```

```python
#### Define a custom function to plot multiple variables
def bivariate(x,y):
    plt.figure(figsize=(20, 8))
    sns.boxplot(x,y,hue='fueltype',data=car)
    plt.legend(bbox_to_anchor=(1.05, 0.5), loc=2, borderaxespad=0.)
    plt.show()
```

#### Data Cleanning - Check for Duplicates , Check for Outliers , Exploratory Data Analysis 

```python
# Identify the duplicates and clean them.
car["CarName"] = car.CarName.map(lambda x: x.split(" ", 1)[0])
car.CarName = car['CarName'].str.lower()
car['CarName'] = car['CarName'].str.replace('vw','volkswagen')
car['CarName'] = car['CarName'].str.replace('vokswagen','volkswagen')
car['CarName'] = car['CarName'].str.replace('toyouta','toyota')
car['CarName'] = car['CarName'].str.replace('porcshce','porsche')
car['CarName'] = car['CarName'].str.replace('maxda','mazda')
car['CarName'] = car['CarName'].str.replace('maxda','mazda')
car.CarName.unique()
```

```python
# Check the outliers using interquartile range (IQR)
car['Log_price'] = np.log(car['price'])
q75, q25 = np.percentile(car.Log_price.dropna(), [75 ,25])
iqr = q75 - q25
min = q25 - (iqr*1.5)
max = q75 + (iqr*1.5)
car['PriceOutlier'] = 0
car.loc[car['Log_price'] < min, 'PriceOutlier'] = 1
car.loc[car['Log_price'] > max, 'PriceOutlier'] = 1
car.groupby('PriceOutlier').count()
```

```python
##### As you can see there's no outliers in the price column and let us remove it.
#### also drop the car unique id column which is of no use for our analysis. 
car.drop(['car_ID'], axis =1, inplace = True)
car.drop(['Log_price'], axis =1, inplace = True)
car.drop(['PriceOutlier'], axis =1, inplace = True)
```

#### Data Preparation - Handling Categorical variables, Applying Normalisation , Dummy Variables and Derived Variables.

```python
# Handling Categorical Variables that are having 2 possible values.
car['fueltype'] = car['fueltype'].map({'gas': 1, 'diesel': 0})
car['aspiration'] = car['aspiration'].map({'std': 1, 'turbo': 0})
car['doornumber'] = car['doornumber'].map({'two': 1, 'four': 0})
car['enginelocation'] = car['enginelocation'].map({'front': 1, 'rear': 0})
```

```python
# For Categorical that are having more values (i.e > 2), let us create dummies using Pandas Libraries.
car = pd.get_dummies(car)
```

```python
# Apply normalisation for Continuous Variables that includes both dependent and independent variables. 
colstonormalize = ['symboling', 'carlength', 'carheight', 
         'enginesize', 'boreratio', 'stroke', 'compressionratio','horsepower', 'peakrpm', 'citympg','price']

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(car[colstonormalize])
car_normalized = pd.DataFrame(x_scaled)

car['symboling'] = car_normalized[0]
car['carlength'] = car_normalized[1]
car['carheight'] = car_normalized[2]
car['enginesize'] = car_normalized[3]
car['boreratio'] = car_normalized[4]
car['stroke'] = car_normalized[5]
car['compressionratio'] = car_normalized[6]
car['horsepower'] = car_normalized[7]
car['peakrpm'] = car_normalized[8]
car['citympg']= car_normalized[9]
car['price']= car_normalized[10]
```

```python
# Let us make the copy of the data for later use for bootstrapping.
car_copy = car.copy()
```

```python
#Split the dataset - 70 and 30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 ,test_size = 0.3, random_state=100)
```

#### Linear Regression -  Apply RFE and remove unwanted features 

```python
#Apply Linear Regression and keep only 20 features.
lmrfe = LinearRegression()
rfe = RFE(lmrfe, 20)             
rfe = rfe.fit(X_train, y_train)
print(rfe.support_)            
print(rfe.ranking_)  
```

```python
#List out the 20 features that are shortlisted by RFE
X_train.columns[rfe.support_]
```

```python
#Eliminate the unwanted features from the train dataset
rfecolremove = X_train.columns[~rfe.support_]
```

```python
#Add the constant
X_train = sm.add_constant(X_train) 
```

The Graph depicts the Actual and predicted and nearly close.

```python
# Plot the error terms
c = [i for i in range(1,63,1)]
fig = plt.figure()
plt.plot(c,y_test-y_pred, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('ytest-ypred', fontsize=16)      
```

## Validate with Bootstrapping

```python
#Create a empty dataframe and bootstrap columns from the copy of the car data set.
car_bs = pd.DataFrame() 
for column in car_copy:
    car_bs[column]=np.random.choice(car_copy[column],size=1500,replace=True)    
```

```python
#take the copy of the actual price value
#drop the columns that are identified during the model build phase. 
#drop dependent feature - price
#add constants
carpriceacutal = car_bs.price
car_bs = car_bs.drop(rfecolremove, axis=1)
car_bs = car_bs.drop(columnsremovedsofar, axis=1)
car_bs = car_bs.drop('price',axis=1)    
car_bs = sm.add_constant(car_bs)
```

```python
#predict the data using bootstrap dataset
carpricepredicted = lm.predict(car_bs)
```

#### Assessing the Model:
Lets scatter plot the error and see if the error has some pattern.

```python
# Actual vs Predicted
c = [i for i in range(1,1501,1)]
fig = plt.figure()
plt.plot(c,carpriceacutal, color="blue", linewidth=3.5, linestyle="-")     #Plotting Actual
plt.plot(c,carpricepredicted, color="red",  linewidth=3.5, linestyle="-")  #Plotting predicted
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Car Price', fontsize=16)  
```

```python
# Error terms
c = [i for i in range(1,1501,1)]
fig = plt.figure()
plt.plot(c,carpriceacutal-carpricepredicted, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('carpriceacutal-carpricepredicted', fontsize=16)    
```
